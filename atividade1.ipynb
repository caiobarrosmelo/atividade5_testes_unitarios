# Casos de Uso - Sistema Calculadora

## UC-01: Realizar Soma

**Descrição:** Permite que o usuário realize a operação matemática de soma.

**Pré-Condições:** Calculadora está operacional

### Fluxo Principal

1. Usuário insere o primeiro número
2. Usuário seleciona operação (+)
3. Usuário insere o segundo número
4. Usuário pressiona "="
5. Sistema calcula e exibe o resultado

### Fluxos Alternativos

- **5a:** Não se aplica à soma
- **5b:** Resultado maior que 100 ou menor que -100 – sistema exibe "Erro: Overflow"
